# Regex practice on a dataset from keggle

```python
import pandas as pd
```

```python
df = pd.read_csv('KaggelMostVotes.csv')
import re
```

```python
df.head(2)
```

Question 1: Find the URLs (Link) that contain "kaggle.com."

```python
df['Link'][df['Link'].str.contains(r'kaggle\.com',regex=True)]
```

Question 2: Extract the file extensions (FileType) from the URLs (Link).

```python
df['FileType'] = df['Link'].str.extract(r'(\.[a-zA-Z]+$)')
df
```

```python
df.FileType.value_counts()
```

Question 3: Identify the datasets with Titles that contain the word "Soccer."

```python
df[df['Title'].str.contains("Soccer.",flags=re.IGNORECASE)]
#IGNORECASE will help us to match to the string soccer irrespective of if it's in lowercase or uppercase it ensures that the search is case-insensitive. We can perform the similar task without the help of re module by simply doing inside the str.contains('soccer',case = False)
```

Question 4: Extract the dataset Uploader names that contain "Machine Learning."

```python
df['Uploader'][df['Uploader'].str.contains('machine learning',case = False)]
```

Question 5: Find the datasets with dataset descriptions (Sub-title) containing the word "fraud."

```python
df[df['Sub-title'].str.contains('fraud',case=False)]
```

Question 6: Extract the file sizes (FileSize) in megabytes (MB) from the dataset information.

```python
a = df['FileSize'].str.extract(r'(\d\.?\d)\sMB').dropna()
a
```

```python
df.FileSize.str.split(' ',expand=True)[1].value_counts()
```

Question 7: Extract the sizes (FileSize) in gigabytes (GB) from the dataset information.

```python
a = df['FileSize'].str.extract(r'(\d\.?\d)\sGB').dropna()
a
```

Question 8: Find the dataset Titles that contain numbers.

```python
df['Title'][df['Title'].str.contains(r'\d+',regex=True)].to_frame()
```

Question 9: Extract the dataset Views values with "k" (thousands) at the end.

```python
df['Views'][df['Views'].str.contains(r'k$',regex=True)].to_frame()
```

Question 10: Identify the dataset Tags that contain the word "finance."

```python
df['Tags'][df['Tags'].str.contains('finance',case=False)].to_frame()
```

Question 11: Find the dataset Links (Link) that are secure URLs (start with "https").

```python
df['Link'][df['Link'].str.contains(r'^https:')]
```

Question 12: Extract the version numbers (Version) with a single digit after the decimal point.

```python
df.Version = df.Version.astype('str')
```

```python
a = df['Version'].str.extract(r'(\d+\.\d{1})').dropna()
```
